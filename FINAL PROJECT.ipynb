## **Context - Terevisi**

Perusahaan telekomunikasi merupakan salah satu pilar penting dalam era digital saat ini, menyediakan berbagai produk dan layanan yang mendukung kebutuhan komunikasi, konektivitas, dan hiburan masyarakat. Layanan yang ditawarkan mencakup telepon seluler, internet broadband, keamanan online, backup data, perlindungan perangkat, dukungan teknis, serta layanan streaming TV dan film. Dengan semakin berkembangnya teknologi, perusahaan telekomunikasi tidak hanya menjadi penyedia layanan dasar, tetapi juga menjadi enabler bagi transformasi digital di berbagai sektor kehidupan. Namun, di balik kemajuan ini, industri telekomunikasi menghadapi tantangan yang kompleks, salah satunya adalah persaingan yang semakin ketat dan perubahan preferensi pelanggan yang dinamis.


Salah satu masalah utama yang dihadapi perusahaan telekomunikasi adalah churn, yaitu fenomena di mana pelanggan memutuskan untuk berhenti menggunakan layanan yang ditawarkan. Tingkat churn yang tinggi dapat berdampak signifikan pada pendapatan perusahaan, mengingat biaya akuisisi pelanggan baru biasanya jauh lebih mahal dibandingkan dengan mempertahankan pelanggan yang sudah ada. Selain itu, churn juga dapat menjadi indikator adanya ketidakpuasan pelanggan terhadap layanan yang diberikan, baik dari segi kualitas, harga, atau pengalaman pengguna. Dalam industri yang sangat kompetitif seperti telekomunikasi, kehilangan pelanggan berarti kehilangan pangsa pasar yang dapat dimanfaatkan oleh pesaing.


Analisis churn menjadi sangat penting karena dapat membantu perusahaan memahami faktor-faktor yang mendorong pelanggan untuk berpindah ke layanan lain. Dengan melakukan analisis ini, perusahaan dapat mengidentifikasi pola dan tren yang berkaitan dengan churn, seperti masalah teknis, ketidakpuasan layanan, atau tawaran yang lebih menarik dari pesaing. Data yang diperoleh dari analisis churn dapat digunakan untuk mengembangkan strategi retensi yang lebih efektif, seperti menawarkan paket layanan yang lebih sesuai, meningkatkan kualitas layanan, atau memberikan insentif kepada pelanggan yang berpotensi churn. Dengan demikian, analisis churn tidak hanya membantu mengurangi tingkat churn, tetapi juga meningkatkan loyalitas pelanggan dan menjaga pertumbuhan bisnis dalam jangka panjang.


Dalam konteks bisnis yang semakin dinamis, analisis churn bukan hanya sekadar alat untuk mempertahankan pelanggan, tetapi juga menjadi kunci untuk memastikan keberlanjutan perusahaan. Tingkat churn yang tinggi dapat mengindikasikan adanya masalah mendasar dalam operasional atau strategi bisnis yang perlu segera diatasi. Oleh karena itu, analisis churn harus dilakukan secara berkala dan komprehensif untuk memastikan perusahaan tetap kompetitif dan responsif terhadap kebutuhan pelanggan. Dengan memahami dan mengatasi masalah churn, perusahaan telekomunikasi tidak hanya dapat mempertahankan pelanggannya, tetapi juga membangun hubungan yang lebih kuat dan berkelanjutan dengan mereka, yang pada akhirnya akan mendorong pertumbuhan dan profitabilitas perusahaan.

**Problem Statement**

Tingginya churn berdampak pada peningkatan biaya akuisisi pelanggan baru dan berkurangnya pendapatan yang stabil. Perusahaan perlu berinvestasi (mengeluarkan biaya) untuk mendapatkan pelanggan baru.  [Pada ulasan Bisnis Hardvard](https://hbr.org/2014/10/the-value-of-keeping-the-right-customers) mengatakan, mendapatkan pelanggan baru bisa 5 sampai 25 kali lebih mahal daripada mempertahankan pelanggan yang sudah ada. Oleh karena itu, untuk mengurangi churn dan meningkatkan retensi pelanggan, perusahaan telekomunikasi perlu mengidentifikasi pelanggan yang berisiko berhenti menggunakan layanan.

**Goals**

Perusahaan ingin  mengidentifikasi pelanggan yang berisiko tinggi berhenti menggunakan layanan perusahaan. Dengan memahami pelanggan yang berisiko churn, perusahaan dapat mengarahkan upaya retensi secara lebih efisien, mengurangi kebutuhan untuk mengeluarkan biaya besar dalam akuisisi pelanggan baru. Dengan menurunkan tingkat churn, perusahaan dapat memastikan aliran pendapatan yang lebih stabil dan meningkatkan profitabilitas jangka panjang.

**Analytic Approach**

Jadi yang akan kita lakukan adalah menganalisis data untuk menemukan pola yang membedakan pelanggan yang beresiko berhenti menggunakan layanan perusahaan dan yang tidak berhenti.

Kemudian kita akan membangun model klasifikasi yang akan membantu perusahaan untuk dapat memprediksi probabilitas pelanggan yang akan berhenti menggunakan layanan atau tidak.

**Metric Evaluation**


**Confusion Matrix**

Dalam memprediksi pelanggan yang berisiko churn, kita menggunakan **Confusion Matrix**, yang terdiri dari empat kategori utama:

1. **True Positives (TP)**: Pelanggan yang benar-benar churn (berhenti menggunakan layanan) dan berhasil diprediksi churn oleh model.  
2. **True Negatives (TN)**: Pelanggan yang tidak churn (tetap menggunakan layanan) dan diprediksi tidak churn oleh model.  
3. **False Positives (FP)**: Pelanggan yang sebenarnya tidak churn, tetapi model salah memprediksi mereka akan churn (*false alarm*).  
4. **False Negatives (FN)**: Pelanggan yang sebenarnya churn, tetapi model gagal mendeteksinya dan memprediksi mereka tidak churn (*salah deteksi*).  

**Jenis Error dan Dampaknya**

1. **Type 1 Error (False Positive)**  
   - **Penjelasan**: Model memprediksi pelanggan akan churn, padahal sebenarnya mereka tidak churn.  
   - **Konsekuensi**: Sumber daya perusahaan akan terbuang untuk memberikan intervensi (seperti penawaran atau insentif) kepada pelanggan yang sebenarnya tidak membutuhkannya. Hal ini dapat menurunkan efisiensi dan mengurangi margin keuntungan.

2. **Type 2 Error (False Negative)**  
   - **Penjelasan**: Model gagal mendeteksi pelanggan yang sebenarnya berisiko churn.  
   - **Konsekuensi**: Pelanggan yang berisiko churn tidak diberi intervensi yang tepat, sehingga mereka benar-benar berhenti menggunakan layanan. Ini berdampak pada penurunan pendapatan perusahaan secara signifikan dan memaksa perusahaan mengeluarkan biaya tambahan untuk mendapatkan pelanggan baru sebagai pengganti.

**Fokus Utama: Meminimalkan False Negatives (FN)**

Dari dua jenis error di atas, **Type 2 Error (False Negative)** memiliki dampak yang lebih besar terhadap bisnis. Jika pelanggan yang berisiko churn tidak terdeteksi, perusahaan tidak dapat mengambil tindakan untuk mempertahankan mereka. Hal ini mengakibatkan:

- Hilangnya pendapatan dari pelanggan yang berhenti.  
- Biaya tambahan untuk akuisisi pelanggan baru, yang jauh lebih mahal daripada mempertahankan pelanggan lama.  

Oleh karena itu, model yang dibangun harus fokus pada **mengurangi False Negatives** agar pelanggan yang berisiko churn dapat teridentifikasi dengan lebih baik.

**Mengapa Memilih Recall sebagai Metric?**

**Recall** adalah metric yang mengukur seberapa baik model mendeteksi pelanggan yang benar-benar churn. Rumus recall adalah:  






Semakin tinggi nilai recall, semakin banyak pelanggan yang benar-benar berisiko churn berhasil terdeteksi. Dengan memilih recall sebagai metric utama, perusahaan dapat memastikan bahwa pelanggan berisiko tidak terlewatkan, sehingga mereka dapat diberikan intervensi yang tepat waktu. Ini akan membantu:

- Mencegah kehilangan pelanggan yang bernilai tinggi.  
- Menjaga stabilitas pendapatan.  
- Mengurangi kebutuhan akan akuisisi pelanggan baru yang mahal.  

Meminimalkan False Negatives adalah prioritas utama karena pelanggan yang tidak terdeteksi dapat berdampak langsung pada pendapatan perusahaan. Dengan menggunakan **recall** sebagai metric utama, model akan lebih fokus pada mendeteksi pelanggan yang benar-benar berisiko churn, sehingga memungkinkan perusahaan mengambil langkah-langkah yang efektif untuk mempertahankan pelanggan dan meningkatkan profitabilitas jangka panjang.

**Feature Selection**

Untuk mengecek apa saja kolom yang memiliki pengaruh signifikan terhadap target, kami menggunakan chi2 dan pval.

Dalam konteks dataset Telco Churn, uji Chi-Square (chi2) dan p-value (pval) berguna untuk menentukan hubungan antara dua variabel kategori.

Semakin tinggi nilai chi2, semakin besar indikasi adanya hubungan antara variabel.

Sementara pada pval, jika pval < 0.05, maka ada hubungan signifikan antara variabel.
Jika pval ≥ 0.05, maka tidak ada bukti kuat bahwa kedua variabel saling berhubungan.

```python
# Feature Selection (Kategori)
from scipy.stats import chi2_contingency

chi_result = []
for column in df_model.select_dtypes(include=['object']) :
    cont_tabel = pd.crosstab(df_model[column], df_model['Churn'])
    chi2, pval, dof, expected = chi2_contingency(cont_tabel)
    pengaruh = pval < 0.05
    chi_result.append([column, chi2, pval, pengaruh])

chi_df = pd.DataFrame(
    chi_result,
    columns=['Feature, Chi-Square Statistics, P Value, Berpengaruh Terhadap Target(?)'.split(', ')]
)
chi_df
```

Berdasarkan hasil uji chi-square di atas, dapat disimpulkan bahwa semua kolom yang diuji memiliki pengaruh signifikan terhadap target (ditunjukkan dengan nilai P Value yang sangat kecil dan berada di bawah ambang signifikan). Oleh karena itu, semua fitur yang ditampilkan akan dimasukkan ke dalam proses pengujian lebih lanjut.

## **Metode Balancing**

Untuk meningkatkan performa model, terutama dalam mendeteksi kelas minoritas, **metode balancing** akan diterapkan. Teknik ini bertujuan untuk menangani ketidakseimbangan data kelas, yang dapat mencakup:
- **Oversampling**: Menambah jumlah data pada kelas minoritas.
- **Undersampling**: Mengurangi jumlah data pada kelas mayoritas.
- **Menggunakan Algoritma dengan Penanganan Ketidakseimbangan**: Seperti memberikan bobot pada kelas minoritas dalam model.

Langkah ini diharapkan dapat meningkatkan performa recall pada model secara keseluruhan.

```python
# Top 3 Model Berdasarkan Hasil Recall

models = {
    'Logistic Regression' : LogisticRegression(solver = 'liblinear', random_state=7),
    'LGBM': LGBMClassifier(random_state=20),
    'XGB' : XGBClassifier(random_state=20, use_label_encoder=False, eval_metric='logloss')
}
```

Metode Balancing yang Digunakan : </br>
- SMOTE (Synthetic Minority Oversampling Technique)
- ADASYN (Adaptive Synthetic Sampling)
- RandomOverSampler
- RandomUnderSampler
- NearMiss
- CondensedNearestNeighbour

```python

balancing_methods = {
    'SMOTE': SMOTE(random_state=201),
    'ADASYN': ADASYN(random_state=201),
    'RandomOverSampler': RandomOverSampler(random_state=201),
    'RandomUnderSampler': RandomUnderSampler(random_state=201),
    'NearMiss': NearMiss(version=3),
    'CondensedNearestNeighbour': CondensedNearestNeighbour(random_state=201)
    }

namaModel = []
namaBalance = []
recall = []

for model in models.items() :
    for i in balancing_methods.items() :
        pipemodel_ov = Pipeline([
        ('preprocessor', transformer),
        ('resample', i[1]),
        ('model', model[1])
        ])

        pipemodel_ov.fit(X_train, y_train)
        y_pred_ = pipemodel_ov.predict(X_test)

        namaModel.append(model[0])
        namaBalance.append(i[0])
        recall.append(round(recall_score(y_test, y_pred_), 2))

pd.DataFrame({
    'Method' : namaModel,
    'Metode Balancing' : namaBalance,
    'Recall' : recall,
}).sort_values(by='Recall', ascending=False).reset_index(drop=True)
```

1. **Logistic Regression + SMOTE**:
   - Recall tertinggi (**0.81**) di antara semua kombinasi, menjadikannya metode terbaik untuk mendeteksi kelas minoritas.

2. **LGBM + RandomUnderSampler**:
   - Memberikan performa terbaik untuk LGBM dengan recall **0.77**, menunjukkan efektivitas dalam menangani ketidakseimbangan kelas.

3. **XGB + NearMiss**:
   - Recall terbaik untuk XGB sebesar **0.75**, menjadikannya metode balancing yang optimal untuk model ini.

## **Hyperparameter Tuning**

Hyperparameter Tuning adalah proses mencari kombinasi **parameter terbaik** untuk sebuah model *machine learning* agar performanya optimal pada dataset tertentu. </br>

Hyperparameter Tuning bertujuan untuk:</br>
1. **Meningkatkan performa model** dengan mengoptimalkan parameter.
2. **Mengurangi overfitting atau underfitting** agar model dapat generalisasi dengan baik.
3. **Menyesuaikan model dengan dataset** untuk hasil prediksi yang optimal.
4. **Maksimalkan efisiensi** tanpa mengganti algoritma yang digunakan.

### **1. Logistic Regression**

```python
# Menggunakan RandomizedSearchCV untuk Efisiensi

from sklearn.model_selection import RandomizedSearchCV
```

```python

# Pipeline
pipeline = Pipeline([
    ('preprocess', transformer),
    ('SMOTE', SMOTE(random_state=201)),
    ('model', logreg)
])

# Hyperparameters yang disederhanakan
hyperparameters = {
    'model__penalty': ['l1', 'l2'],  # Fokus pada penalty umum
    'model__C': [0.1, 1, 10],        # Hilangkan nilai ekstrem
    'model__solver': ['liblinear', 'saga'],  # Kompatibel dengan l1 dan l2
    'model__max_iter': [500, 1000]   # Iterasi yang cukup untuk konvergensi
}

# Menggunakan RandomizedSearchCV untuk efisiensi
grid_logreg = RandomizedSearchCV(
    pipeline,
    hyperparameters,
    n_iter=10,  # Batasi iterasi
    cv=3,       # Kurangi lipatan cross-validation
    scoring='recall',
    random_state=201,
    n_jobs=-1
)

# Melatih model
grid_logreg.fit(X_train, y_train)

# Mendapatkan parameter terbaik
best_params = grid_logreg.best_params_
```

```python
print('Tabel berikut menunjukkan hasil dari grid search atau hyperparameter tuning untuk model Logistic Regression, di mana kombinasi terbaik dari parameter telah ditemukan.')
best_params_table = pd.DataFrame([grid_logreg.best_params_])
best_params_table
```

### **2. LGBM**

```python


# Pipeline
pipeline = Pipeline([
    ('preprocess', transformer),
    ('randUnder', RandomUnderSampler(random_state=20)),
    ('model', lgbm)
])

# Hyperparameters yang disederhanakan
hyperparameters = {
    'model__n_estimators': [100, 200],  # Fokus pada nilai umum yang optimal
    'model__max_depth': [10, 20],       # Hilangkan -1 untuk menghindari pohon tak terbatas
    'model__num_leaves': [31, 50],      # Cukup dua opsi untuk mengurangi kombinasi
    'model__min_child_samples': [20, 50],
    'model__learning_rate': [0.05, 0.1],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0],
    'model__reg_lambda': [1, 10],
    'model__scale_pos_weight': [1, sum(y_train == 0) / sum(y_train == 1)]
}

# Menggunakan RandomizedSearchCV untuk efisiensi
grid_lgbm = RandomizedSearchCV(
    pipeline,
    hyperparameters,
    n_iter=20,  # Batasi jumlah iterasi
    cv=3,       # Kurangi cross-validation ke 3 lipatan
    scoring='recall',
    random_state=20,
    n_jobs=-1
)

# Melatih model
grid_lgbm.fit(X_train, y_train)

# Mendapatkan parameter terbaik
best_params = grid_lgbm.best_params_
```

```python
print('Tabel berikut menunjukkan hasil dari grid search atau hyperparameter tuning untuk model LGBM, di mana kombinasi terbaik dari parameter telah ditemukan.')
best_params_table = pd.DataFrame([grid_lgbm.best_params_])
best_params_table
```

```python
from sklearn.metrics import recall_score

estimator_logreg_default = Pipeline([
    ('preprocess', transformer),
    ('SMOTE', SMOTE(random_state=201)),
    ('model', logreg)
])

estimator_logreg_default.fit(X_train, y_train)

logreg_tuned = LogisticRegression(
    C=10,               # Parameter regularisasi
    max_iter=1000,      # Jumlah iterasi maksimum
    penalty='l1',      # Penalti L1 (Ridge Regularization)
    solver='liblinear' # Solver untuk optimasi
)

estimator_logreg_tuned=Pipeline([
    ('preprocess', transformer),
    ('SMOTE', SMOTE(random_state=201)),
    ('model', logreg_tuned)
])

estimator_logreg_tuned.fit(X_train, y_train)

y_pred_default = estimator_logreg_default.predict(X_test)
y_pred_tuned = estimator_logreg_tuned.predict(X_test)

recall_score_default = recall_score(y_test, y_pred_default)
recall_score_tuned = recall_score(y_test, y_pred_tuned)

print('Recall Score Default Logistic Regression : ', recall_score_default)
print('Recall Score Tuned Logistic Regression : ', recall_score_tuned)
```
